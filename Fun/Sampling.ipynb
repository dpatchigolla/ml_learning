```python
import numpy as np
```

```python
P=np.random.beta(a=20,b=500,size=100000)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ax=sns.distplot(P)
ax.set_xbound(0,0.15)
```

```python
np.mean(P)
```

```python
np.var(P)
```

```python
samp_size=10000
```

```python
P1=np.random.choice(P, size=samp_size)
[np.mean(P1), np.var(P1)]
```

```python
P2=np.random.choice(P, size=samp_size, replace=False)
[np.mean(P2), np.var(P2)]
```
