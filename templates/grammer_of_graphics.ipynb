```python
from plotnine import *
from plotnine.data import mtcars
```

```python
mtcars.head()
```

```python
mtcars.shape
```

## 2-D Visualization

```python
(ggplot(mtcars, 
      aes('wt','mpg'))
+geom_point()
+theme_bw())
```

```python
(ggplot(mtcars, 
      aes('wt','mpg'))
+geom_point())
```

```python
(ggplot(mtcars, 
      aes('wt','mpg'))
+geom_point()
+theme_seaborn())
```

```python
(ggplot(mtcars, 
      aes('wt','mpg'))
+geom_point()
+theme_xkcd())
```

### 3-D scatter plot

```python
(ggplot(mtcars,
       aes('wt','mpg',color='factor(gear)'))
       +geom_point()
       +theme_538())
```

## 4-D scatter plot with color and size

```python
(ggplot(mtcars, 
       aes('wt', 'mpg', color='factor(gear)', size='cyl'))
+geom_point()
+theme_xkcd())
```

## Facet grid - plotting multiple charts together

```python
(ggplot(mtcars, 
       aes('wt','mpg',color='factor(gear)'))
+geom_point()
+facet_wrap('cyl')
+theme_minimal())
```

## Adding Trend lines 

```python
(ggplot(mtcars, aes('wt','mpg',
                   color='factor(gear)'))
+geom_point()
+stat_smooth(method='glm')
+theme_538())
```

## Visualize 5 Dimensions
(x,y,color, size, facet)

```python
(ggplot(mtcars, aes('wt','mpg', 
                   color='factor(gear)', size='hp'))
+geom_point()
+facet_wrap('cyl')
)
```

## Facets in both x and y axes

```python
(ggplot(mtcars, aes('wt','mpg', 
                   color='factor(gear)', size='hp'))
+geom_point()
+facet_grid('cyl~am')
)
```
